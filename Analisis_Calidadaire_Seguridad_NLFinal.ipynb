```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df1 = pd.read_csv('datos.csv', usecols=['Location','Excelente o Buena','Regular','Mala','Pesima'])
df2 = pd.read_csv('datos.csv',usecols=['Location','Muy seguro','Seguro','Inseguro','Muy inseguro'])
cal_aire1 = df1['Excelente o Buena']
cal_aire2 = df1['Regular']
cal_aire3 = df1['Mala']
cal_aire4= df1['Pesima']
seg1 = df2['Muy seguro']
seg2 = df2['Seguro']
seg3 = df2['Inseguro']
seg4 = df2['Muy inseguro']
loc = df1['Location']
```

# Analisis Descriptivo

```python
#1. Estadisticos de Opinion en Calidad de aire
aire_media1 = np.mean(cal_aire1)
aire_media2 = round(np.mean(cal_aire2), 2)
aire_media3 = round(np.mean(cal_aire3), 2)
aire_media4 = round(np.mean(cal_aire4), 2)
print('Media por Estado:\n')
print('Media de Excelente o Buena: ', aire_media1,  '\nMedia de Regular:', aire_media2,  '\nMedia de Mala: ', aire_media3,  '\nMedia de Pesima: ', aire_media4 , '\n')

aire_mediana1 = round(np.median(cal_aire1), 2)
aire_mediana2 = round(np.median(cal_aire2), 2)
aire_mediana3 = round(np.median(cal_aire3), 2)
aire_mediana4 = round(np.median(cal_aire4), 2)
print('Mediana por Estado: \n')
print('Mediana de Excelente o Buena: ', aire_mediana1,  '\nMediana de Regular:', aire_mediana2,  '\nMediana de Mala: ', aire_mediana3,  '\nMediana de Pesima: ', aire_mediana4 , '\n')

aire_moda1 = round(stat.mode(cal_aire1), 2)
aire_moda2 = round(stat.mode(cal_aire2), 2)
aire_moda3 = round(stat.mode(cal_aire3), 2)
aire_moda4 = round(stat.mode(cal_aire4), 2)
print('Moda por Estado:\n')
print('Moda de Excelente o Buena: ', aire_moda1,  '\nModa de Regular:', aire_moda2,  '\nModa de Mala: ', aire_moda3,  '\nModa de Pesima: ', aire_moda4 , '\n')

aire_desvest1 = round(stat.pstdev(cal_aire1), 2)
aire_desvest2 = round(stat.pstdev(cal_aire2), 2)
aire_desvest3 = round(stat.pstdev(cal_aire3), 2)
aire_desvest4 = round(stat.pstdev(cal_aire4), 2)
print('Desviacion Estandar por Estado:\n')
print('Desviacion Estandar de Excelente o Buena: ', aire_desvest1,  '\nDesviacion Estandar de Regular:', aire_desvest2,  '\nDesviacion Estandar de Mala: ', aire_desvest3,  '\nDesviacion Estandar de Pesima: ', aire_desvest4 , '\n')

```

```python
#1. Estadisticos de Opinion en Seguridad
seg_media1 = round(np.mean(seg1), 2)
seg_media2 = round(np.mean(seg2), 2)
seg_media3 = round(np.mean(seg3), 2)
seg_media4 = round(np.mean(seg4), 2)
print('Media por Estado:\n')
print('Media de Muy seguro: ', seg_media1,  '\nMedia de Seguro:', seg_media2,  '\nMedia de Inseguro: ', seg_media3,  '\nMedia de Muy Inseguro: ', seg_media4 , '\n')

seg_mediana1 = round(np.median(seg1), 2)
seg_mediana2 = round(np.median(seg2), 2)
seg_mediana3 = round(np.median(seg3), 2)
seg_mediana4 = round(np.median(seg4), 2)
print('Mediana por Estado: \n')
print('Mediana de Muy seguro: ', seg_mediana1,  '\nMediana de seguro:', seg_mediana2,  '\nMediana de inseguro: ', seg_mediana3,  '\nMediana de Muy inseguro: ', seg_mediana4 , '\n')

seg_moda1 = round(stat.mode(seg1), 2)
seg_moda2 = round(stat.mode(seg2), 2)
seg_moda3 = round(stat.mode(seg3), 2)
seg_moda4 = round(stat.mode(seg4), 2)
print('Moda por Estado:\n')
print('Moda de Muy seguro: ', seg_moda1,  '\nModa de Seguro:', seg_moda2,  '\nModa de Inseguro: ', seg_moda3,  '\nModa de Muy inseguro: ', seg_moda4 , '\n')

seg_desvest1 = round(stat.pstdev(seg1), 2)
seg_desvest2 = round(stat.pstdev(seg2), 2)
seg_desvest3 = round(stat.pstdev(seg3), 2)
seg_desvest4 = round(stat.pstdev(seg4), 2)
print('Desviacion Estandar por Estado:\n')
print('Desviacion Estandar de Muy seguro: ', seg_desvest1,  '\nDesviacion Estandar de Seguro:', seg_desvest2,  '\nDesviacion Estandar de Inseguro: ', seg_desvest3,  '\nDesviacion Estandar de Muy inseguro: ', seg_desvest4 , '\n')
```

# Coeficiente de Correlacion

```python
#Calcular coeficiente de correlacion
coef_corr1 = np.corrcoef(x1,y1)[0,1]
coef_corr2 = np.corrcoef(x2,y2)[0,1]
coef_corr3 = np.corrcoef(x3,y3)[0,1]
coef_corr4 = np.corrcoef(x4,y4)[0,1]
print("Coeficiente de correlacion entre el nivel 1 de calidad de aire y seguridad: ", coef_corr1, "\nCoeficiente de correlacion entre el nivel 2 de calidad de aire y seguridad: ", coef_corr2, "\nCoeficiente de correlacion entre el nivel 3 de calidad de aire y seguridad: ", coef_corr3,"\nCoeficiente de correlacion entre el nivel 4 de calidad de aire y seguridad: ", coef_corr4)
```

# Analisis de Regresion

```python
X=x1.values.reshape(-1,1)
y=y1

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)

plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresi√≥n Lineal')
plt.show()
```
